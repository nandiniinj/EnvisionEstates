## Train Test Splitting

```python
#for learning purpose
#now this function is already there in scikit so we can import it
import numpy as np
def split_train_test(data,test_ratio):
    np.random.seed(42) 
    shuffled=np.random.permutation(len(data))
    test_set_size=int(len(data)*test_ratio)
    test_indices=shuffled[:test_set_size]
    train_indices=shuffled[test_set_size:]
    return data.iloc[train_indices],data.iloc[test_indices]
```

## Missing attributes

To take care of the missing attributes, you have 3 options:
1. get rid of the missing data points (not feasible if there are many missing values).
2. get rid of the whole attribute (not feasible if the attribute is important, here RM is an important attribute because it is a strong positive correlation).
3. set the value to some value(0, mean or median).

```python
a= housing.dropna(subset=['RM']) #option 1
a.shape
#note that the original housing dataframe will remain unchanged
```

```python
housing.drop('RM',axis=1) #option 2
#RM column deleted
#note that the original housing dataframe will remain unchanged
```

```python
housing.drop('RM',axis=1).shape #option 2
```

```python
median=housing['RM'].median() #option 3
housing['RM'].fillna(median)
#note that the original housing dataframe will remain unchanged
```

```python
housing.shape
```

```python
#before we started filling missing attributes
housing.describe()
#RM has less value i.e. 401
```

```python
# automating the above process using Imputer

from sklearn.impute import SimpleImputer
imputer = SimpleImputer(strategy='median') #strategy is a hyperparameter
imputer.fit(housing) #fitting into the dataset
```

```python
#the values it calculated for the median
imputer.statistics_
```

```python
imputer.statistics_.shape #calculated for all 15 values
```

```python
X= imputer.transform(housing)
```

```python
housing_tr= pd.DataFrame(X, columns=housing.columns)
#housing_tr is the transformed dataset which has no missing values now
```

```python
housing_tr.describe()
#now RM is 404, hence the missing values has been filled
```

## Scikit-learn Design

Scikit learn is a library

Primarily, three types of objects:

1. Estimators - It estimates some parameter based on a dataset. Eg.: imputer
     It has a fit method and a transform method.
     Fit method- Fits the dataset and calculates internal parameters.

2. Transformers - transform method takes input and returns output based on the learnings from fit().
        It also has a convenience function called fit_transform() which fits and then transforms.

3. Predictors - LinearRegression model is an example of predictor.
        fit() and predict() are two common functions. 
        It also gives score function which will evaluate the predictions.

## Feature Scaling

Primarily, two types of feature scaling methods:


1. Min-max scaling (Normalization)

    (value-min)/(max-min)
    
    the values lie between 0 to 1.
    Sklearn provides a class called MinMaxScaler for this.
    
    
    

2. Standardization

    (value-mean)/standard deviation
    
    Sklearn provides a class called Standard Scaler for this.

## Creating a Pipeline

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
#pipeline takes a list 
my_pipeline= Pipeline([('imputer', SimpleImputer(strategy="median")), 
                       #... add as many as you want.
                      ('std_scaler', StandardScaler()),
                      ])
```

```python
housing_num_tr= my_pipeline.fit_transform(housing)
```
